The total probability of an almost consecutive permutation occurring for various $n$
=============================================

In [this paper](https://arxiv.org/abs/1907.07172), we show that in a random symmetric walk, the ordinal pattern probability of an almost consecutive permutation does not depend upon choice of symmetric density function. (Almost consecutive permutations are defined in the paper as those whose consecutive values are at most two positions apart in the $1$-line notation.) This notebook adds all of these probabilities for various $n$. The result is a Bernoulli trial that works for a large class of density functions.

```python
%run ../calculations.py
```

```python
{n:almost_consecutive_bernoulli_fraction(n) for n in range(2,9)}
```

A random walk whose steps are generated by an asymmetric density function might still have these ordinal pattern probabilities. It is nice that these probabilities are high enough to be detected in a small sample. By contrast, the number of almost consecutive permutations is much less frequent, as shown below.

```python
from itertools import permutations
{n:len([p for p in permutations(range(1,n+1)) if is_almost_consecutive(p)]) for n in range(2,9)}
```

Thus, a random walk of $8$ symmetric steps produces an almost consecutive permutation with probability near $1/10$. By contrast, an independent random source produces ordinal patterns uniformly. When $8$ consecutive data points are produced by an independent random source, the resulting sequence is almost consecutive with probability $88/40320$, which is less than $1/400$.
